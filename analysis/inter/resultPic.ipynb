```python
%matplotlib inline
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns

param = 'linear_weights.npy'
path = "./DLinearSingle_4_MinMax/"

linear_weights = np.load(os.path.join(path,param))
# 计算权重的绝对值
abs_linear_weights = linear_weights#np.abs(linear_weights)

# 创建一个图形对象和轴对象
plt.figure(figsize=(10, 8))

# 使用 seaborn 绘制热力图
sns.heatmap(abs_linear_weights, annot=False, cmap='viridis', square=True)

# 添加标题和标签
plt.title('Heatmap of Absolute Linear Weights')
plt.xlabel('Features')
plt.ylabel('Output Neurons')

# 显示热力图
plt.savefig('./pics/DLinearSingle_4_MinMax.png')
```
