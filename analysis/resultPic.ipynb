```python
%matplotlib inline
import numpy as np
import os
import matplotlib.pyplot as plt

param = 'long_term_forecast_test_LightTime_custom_ftS_sl192_ll48_pl36_dm128_nh4_el2_dl1_df2048_fc1_ebtimeF_dtTrue_test_0'
path = "../results/"

trues = np.load(os.path.join(path,param,'true.npy'))
trues = trues[::36,:,:]
trues = trues.reshape(-1)

preds = np.load(os.path.join(path,param,'pred.npy'))
preds = preds[::36,:,:]
preds = preds.reshape(-1)

print(trues.shape)

# 
fig, ax = plt.subplots(figsize=(20,8))

# trues
ax.plot(trues, color='red', label='trues')

# preds
ax.plot(preds, color='blue', label='preds')

# # 36
# for i in range(0, len(trues), 36):
#     ax.axvline(x=i, color='gray', linestyle='dashed')

# 
ax.legend()

# 
ax.set_xlabel('Time Point')
ax.set_ylabel('Wind Speed')
ax.set_title('Wind Speed Forecasting')
plt.savefig('./resultPic/Patch_dm128_nh4_st18_pt72_withouNorma_lsl192.png',bbox_inches='tight')
```
